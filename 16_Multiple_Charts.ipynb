```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 깨짐 현상 해결
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 기호 꺠짐 현상 해결
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('여러 그래프 넣기')

# 첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어 점수')
axs[0, 0].set_title('첫 번째 그래프')
axs[0, 0].legend(loc='upper center')
axs[0, 0].set(xlabel='이름', ylabel='점수')
axs[0, 0].set_facecolor('lightyellow')
axs[0, 0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학 점수')
axs[0, 1].plot(df['이름'], df['영어'], label='영어 점수')
axs[0, 1].set_title('두 번째 그래프')
axs[0, 1].legend(loc='upper center')

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'], label='키')

# 네 번째 그래프
axs[1, 1].plot(df['이름'], df['사회'], label='사회 점수', marker='o', color='green', alpha=0.5)
```
