```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 깨짐 현상 해결
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 기호 꺠짐 현상 해결
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend()
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc='lower right')
```

```python
plt.plot(x, y, label='범례')
plt.legend(loc='best')
```

```python
plt.plot(x, y, label='범례')
plt.legend(loc=(0.5,0.5))
```
