```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 깨짐 현상 해결
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 기호 꺠짐 현상 해결
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]

plt.bar(labels, values)
```

```python
plt.bar(labels, values, color='r')
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]
colors = ['r','g','b']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
plt.bar(labels, values, color=colors, alpha=0.5)
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, color=colors, alpha=0.5, width=0.5)
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, color=colors, alpha=0.5, width=0.5)
plt.ylim(175, 195)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
```

```python
ticks = ['1번','2번','3번']

plt.bar(labels, values)
plt.xticks(labels, ticks)
```
