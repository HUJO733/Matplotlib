```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 깨짐 현상 해결
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 기호 꺠짐 현상 해결
```

```python
values = [30,25,20,13,10,2]
labels = ['Python','Java','JavaScript','C#','C/C++','ETC']

plt.pie(values, labels=labels, autopct='%.1f')
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90)
plt.show()
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
explode = [0.2, 0.1, 0, 0, 0, 0]

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
explode = [0.05] * 6

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
explode = [0.05] * 6

plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
explode = [0.05] * 6

plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3), title='언어별 선호도')
plt.show()
```

```python
explode = [0.05] * 6

plt.title('언어별 선호도')

plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3), title='언어')
plt.show()
```
