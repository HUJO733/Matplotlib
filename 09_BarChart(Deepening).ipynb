```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 깨짐 현상 해결
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 기호 꺠짐 현상 해결
```

```python
labels = ['강백호','서태웅','정대만']
values = [190,187,184]

plt.barh(labels, values)
```

```python
plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('*')
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='g')
```
