```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글 깨짐 현상 해결
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 기호 꺠짐 현상 해결

x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
```

```python
days = [1,2,3] # 1일, 2일, 3일
az = [2,4,8] 
pfizer = [5,1,3]
moderna = [1,2,5]

plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', ls='--')
plt.plot(days, moderna, label='moderna', marker='v', ls=':')

plt.legend()
```
